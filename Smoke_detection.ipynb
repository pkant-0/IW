````python
import RPi.GPIO as GPIO

# GPIO pin numbers for smoke and heat detectors
SMOKE_PIN = 17
HEAT_PIN = 18

# GPIO pin number for the hooter alarm
HOOTER_PIN = 27

def initialize():
    GPIO.setmode(GPIO.BCM)

    # Set up smoke detector pin as input
    GPIO.setup(SMOKE_PIN, GPIO.IN)

    # Set up heat detector pin as input
    GPIO.setup(HEAT_PIN, GPIO.IN)

    # Set up hooter alarm pin as output
    GPIO.setup(HOOTER_PIN, GPIO.OUT)
    GPIO.output(HOOTER_PIN, GPIO.LOW)

def smoke_callback(channel):
    # Smoke detection callback function
    if GPIO.input(channel):
        print("Smoke detected! Alerting firefighters.")
        GPIO.output(HOOTER_PIN, GPIO.HIGH)
    else:
        print("Smoke cleared.")
        GPIO.output(HOOTER_PIN, GPIO.LOW)

def heat_callback(channel):
    # Heat detection callback function
    if GPIO.input(channel):
        print("Heat detected! Alerting firefighters.")
        GPIO.output(HOOTER_PIN, GPIO.HIGH)
    else:
        print("Heat cleared.")
        GPIO.output(HOOTER_PIN, GPIO.LOW)

def main():
    initialize()

    # Set up interrupt handlers for smoke and heat detectors
    GPIO.add_event_detect(SMOKE_PIN, GPIO.BOTH, callback=smoke_callback)
    GPIO.add_event_detect(HEAT_PIN, GPIO.BOTH, callback=heat_callback)

    try:
        while True:
            # Your main program logic can go here
            pass

    except KeyboardInterrupt:
        GPIO.cleanup()

if __name__ == '__main__':
    main()
````
